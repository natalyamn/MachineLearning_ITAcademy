### Ejercicio 1. Consumir una API
**Selecciona una API pública y extrae datos utilizando Python.**

```python
import requests
import json 

#API seleccionada: Cat Facts
URL = "https://cat-fact.herokuapp.com/facts"
response = requests.get(URL)

if response.status_code == 200:
    print('Solicitud exitosa. Se han recuperado estos hechos interesantes sobre los gatos:', end="\n"*2)
    facts = response.json()

    #recupero únicamente el texto del hecho
    i = 0    
    for fact in facts:
        i += 1
        print(f"{i}: {fact['text']}", end="\n"*2)
        
else:
    print('Error en la solicitud, detalles:', response.text)
```

### Ejercicio 2. Obtener datos con Web Scraping
**Selecciona un sitio web y extrae datos utilizando la técnica de Web Scraping en Python.**

Web Seleccionada: Top Películas de 2024 (FilmAffinity)

¿Qué elementos puedo extraer?
* Título
* Director

¿Qué elementos no voy a extraer?
* Año: redundante
* Género: no todas las películas lo tienen descrito (ej: Dune)
* Elenco: no todas las películas lo tienen descrito (ej: Memorias de un caracol)


Referencia:



```python
from bs4 import BeautifulSoup
import requests

URL = "https://www.filmaffinity.com/es/ranking.php?rn=ranking_2024_topmovies"
response = requests.get(URL)

soup = BeautifulSoup(response.content, "html.parser")

#selecciono la parte de código que contiene información sobre películas
movie_elements = soup.find_all('div', class_='movie-card mc-flex movie-card-0')

i = 0
for movie in movie_elements:
    i += 1
    title = movie.find('div', class_='mc-title').a['title']
    director = movie.find('div', class_='mc-director').find('span', class_='nb').a['title']

    print(f'{i}: "{title}", dirigida por {director}.')
```
